````python
ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}

list = [(k, v) for k, v in ContinentDict.items()]
````

````python
#display(list)
Continentdf = pd.DataFrame(list)

Continentdf.rename(columns = {1 : 'Continent', 0: 'Country'},
                 inplace = True)

#Continentdf= Continentdf.groupby(["Continent"]).count()
# joined_2 = pd.merge(joined_1, ScimEn , how = 'inner' ,   on='Country')

top_pop=  pd.merge(top_15 , Continentdf , how ='inner' , on = 'Country' )
top_pop["Popoulation"]= top_pop["Energy Supplys"] / top_pop["Energy Supply per Capita"]



display(top_pop)
````

````python
#top_pop= top_pop.groupby(["Continent"]).count()
#sales_df.groupby("Product_Category")['UnitPrice(USD)'].mean().reset_index()

agg_df1=top_pop.groupby("Continent")["Popoulation"].agg([np.sum, np.mean, np.std]).reset_index()

agg_df2=top_pop.groupby("Continent")["Country"].count()

agg_df3= pd.merge(agg_df2 , agg_df1 , how = 'inner' , on= 'Continent' )
agg_df3.rename(columns = {'Country' : 'Size'},
                 inplace = True)
agg_df3= agg_df3.set_index('Continent')
display(agg_df3)
````
